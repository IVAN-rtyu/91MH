```python
import pandas as pd
import numpy as np
import os
import xlrd
import matplotlib
from matplotlib import pyplot as plt
```

```python
R = 0.0136
r = 0.012575
M = 0.000719
d = 0.0252
D = 0.0272
L = 0.05037

path = "data"
```

```python
tmp_df = []

for name in os.listdir(path):
    if "XLS" not in name:
        continue
    wb = xlrd.open_workbook(f"{path}/{name}", encoding_override='iso-8859-1')
    df = pd.read_excel(wb)
    df.columns = df.iloc[3].to_list()
    df = df.iloc[5:-1, 1:-1].reset_index(drop=True).astype(float)
    df['Torque'] /= 100
    df.insert(1, "Angular velocity", (df.iloc[:, 1] * 2 * np.pi) / 60)
    df.loc[:, "Shear Stress"] = M * df.loc[:, "Torque"] / (2 * np.pi * r ** 2 * L)
    df.loc[:, "Shear Rate"] = r ** 2 * df.loc[:, "Angular velocity"] / (R ** 2 - r ** 2)
    df.loc[:, "Viscosity"] = df.loc[:, "Shear Stress"] / (2 * df.loc[:, "Shear Rate"])
#     df = df.groupby(" Speed").max().reset_index()
    tmp_df.append(df)
    display(df)
    df.to_csv(f"{path}/{name[:-4]}.csv")
```

```python
df_all = pd.DataFrame(columns=df.columns)

for i in range(len(tmp_df)):
    df_all.loc[i] = [list(tmp_df[i][j]) if " C" not in j else tmp_df[i][j][0] for j in tmp_df[i].columns]

df_all.to_csv(f"{path}/all_data.csv")
display(df_all)
```

```python
matplotlib.rcParams.update({'font.size': 14})

plt.figure(figsize=(20, 12))
legend_list = []
for i in range(df_all.shape[0]):
    plt.scatter(df_all.loc[i, "Shear Rate"], df_all.loc[i, "Viscosity"], s=50, alpha=.5)
    plt.xlabel("Shear rate")
    plt.ylabel("Viscosity")
    plt.ylim([0.032, 0.043])
    plt.legend([1, 2, 3, 4, 5])
plt.show
```
